## Tugas praktikum

Coba dengan beberapa parameter lain:
- Ubah jumlah neuron di hidden layer (misal: 256 dan 128).
- Tambahkan satu hidden layer lagi.
- Bandingkan akurasi dan waktu pelatihan.
- Eksperimen dengan fungsi aktivasi Sigmoid vs ReLU.

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Bangun model JST
model = Sequential([
    Flatten(input_shape=(28, 28)), # Ubah gambar 28x28 menjadi vektor
    Dense(128, activation='relu'), # Hidden layer 1
    Dense(64, activation='relu'),  # Hidden layer 2
    Dense(10, activation='softmax') # Output layer (10 kelas)
])

# 3. Kompilasi model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 4. Latih model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1)

# 5. Evaluasi model
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi pada data uji: {acc:.4f}")
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import time
import pandas as pd

# -----------------------------
# 1. Load dataset MNIST
# -----------------------------
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# -----------------------------
# 2. Fungsi membuat model
# -----------------------------
def build_model(activation="relu", neurons1=256, neurons2=128, add_layer=False):
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    model.add(Dense(neurons1, activation=activation))
    model.add(Dense(neurons2, activation=activation))

    # Tambah hidden layer ketiga (opsional)
    if add_layer:
        model.add(Dense(neurons2, activation=activation))

    model.add(Dense(10, activation="softmax"))
    return model

# -----------------------------
# 3. Daftar skenario eksperimen
# -----------------------------
experiments = [
    ("ReLU 256-128", "relu", False),
    ("ReLU 256-128 + Extra Layer", "relu", True),
    ("Sigmoid 256-128", "sigmoid", False),
    ("Sigmoid 256-128 + Extra Layer", "sigmoid", True)
]

results = []

# -----------------------------
# 4. Loop menjalankan eksperimen
# -----------------------------
for name, activation, extra_layer in experiments:
    print(f"\n=== Menjalankan model: {name} ===")

    model = build_model(activation=activation,
                        neurons1=256,
                        neurons2=128,
                        add_layer=extra_layer)

    model.compile(optimizer="adam",
                  loss="categorical_crossentropy",
                  metrics=["accuracy"])

    start = time.time()

    history = model.fit(X_train, y_train,
                        epochs=5,
                        batch_size=128,
                        verbose=0)

    duration = time.time() - start

    loss, acc = model.evaluate(X_test, y_test, verbose=0)

    results.append([name, acc, duration])

# -----------------------------
# 5. Tampilkan hasil dalam tabel
# -----------------------------
df = pd.DataFrame(results, columns=["Model", "Akurasi", "Waktu Latih (detik)"])
print("\n===== HASIL PERBANDINGAN =====")
print(df)
```
