```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('drive/MyDrive/ML/wbc.csv')
print(df.columns)

#  Hapus hanya kolom yang tidak digunakan
df = df.drop(columns=['Unnamed: 32'], errors='ignore')
print(df['diagnosis'].unique())

# encoding jika nilainya 'M' dan 'B'
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

# Standarisasi semua kolom numerik kecuali diagnosis
scaler = StandardScaler()
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.drop('diagnosis')
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print(df.head())
```
